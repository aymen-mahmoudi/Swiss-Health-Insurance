```python
import numpy as np
import matplotlib.pyplot as plt

PR300 = 400
PR2500 = 300


cout_ann_supp = (PR300 - PR2500) * 12

print("Cout annuel supplementaire:", cout_ann_supp )

def calc_remboursement(dep, FR):
    rem = 0
    if dep > FR:
        rem = (dep - FR) * 0.9
    return rem


x = np.linspace(0, 2500, 250)
y = [calc_remboursement(value, PR300) for value in x]

plt.axhline(y=cout_ann_supp, color='r', linestyle='--')

plt.plot(x, y)
plt.xlabel("dépenses")
plt.ylabel("Remboursement")
plt.grid()
plt.show()
```
